```python
def skew(seq):
    sk = [0]
    for char in seq:
        if char == "G":
            sk.append(sk[-1]+1)
        elif char == "C":
            sk.append(sk[-1]-1)
        else:
            sk.append(sk[-1])

    return sk

if False:
    with open("dataset_7_10.txt", "r") as f:
        data = f.read()

    seq = [
        "GCATACACTTCCCAGTAGGTACTG",
        "ACCG",
        "ACCC"
    ]

    for s in seq:
        print(skew(s))
```
